````python
import matplotlib.pyplot as plt
import gc

def show_mask(mask, ax, random_color=False):
    if random_color:
        color = np.concatenate([np.random.random(3), np.array([0.6])], axis=0)
    else:
        color = np.array([30 / 255, 144 / 255, 255 / 255, 0.6])
    h, w = mask.shape[-2:]
    mask_image = mask.reshape(h, w, 1) * color.reshape(1, 1, -1)
    ax.imshow(mask_image)
    del mask
    gc.collect()

def show_masks_on_image(raw_image, outputs):
  plt.imshow(np.array(raw_image))
  ax = plt.gca()
  ax.set_autoscale_on(False)
  for item in outputs:
      show_mask(item["segmentation"], ax=ax, random_color=True)
  plt.axis("off")
  plt.show()
  gc.collect()
````

````python
# compute similarity score for two images based on their segmentation masks
def compare_image_masks(image_a_outputs, image_b_outputs):
    largest_num_masks = max(len(image_a_outputs), len(image_b_outputs))

    cumulative_score = 0
    for index_a, mask_a in enumerate(image_a_outputs):
        best_match = 0

        compare_two_masks = lambda mask_b: (mask_a["segmentation"] & mask_b["segmentation"]).sum() / max(mask_a["area"], mask_b["area"])
        v_compare_two_masks = np.vectorize(compare_two_masks)

        results = v_compare_two_masks(image_b_outputs)
        best_match = max(results)
        
        cumulative_score += best_match

    return cumulative_score / largest_num_masks

image_a_outputs = screenshots[0]
image_b_outputs = screenshots[1]
print(compare_image_masks(image_a_outputs, image_b_outputs)) # Should be high, similar screenshots and masks
````

````python
from multiprocessing import Pool

def wrapped_compare(index_a, index_b):
    return (index_b, compare_image_masks(screenshots[index_a], screenshots[index_b]))

keys = sorted(screenshots.keys())
pairs = [(keys[i], keys[i+1]) for i in range(0, len(keys)-1)]

with Pool() as pool:
    results = pool.starmap(wrapped_compare, pairs)

threshold = 0.6
groups = []
group = [0]
for index, score in results:
    if score > threshold:
        group.append(index)
    else:
        groups.append(group)
        group = [index]
groups.append(group)
````
